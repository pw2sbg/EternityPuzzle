```python
from Puzzle import *
from copy import deepcopy
import time
import matplotlib.pyplot as plt
```

```python
path = "pieces_set/"
file_name = path + "pieces_16x16.txt"
puzzle = Puzzle(file_name=file_name)
max_puzzle_score = puzzle.rows*(puzzle.columns-1) + (puzzle.rows-1)*puzzle.columns
eternity = puzzle.initialize()
```

```python
itr = 30000
init_puzzle = eternity
init_score = eternity.get_score()
iterations, scores = [], []

start = time.time()
for i in range(itr):
  # print(i)
  iterations.append(i)
  scores.append(init_score)
  best_puzzle = deepcopy(init_puzzle)
  tmp_puzzle = deepcopy(init_puzzle)
  swapped_puzzle = tmp_puzzle.swap_piece()
  swapped_score = swapped_puzzle.get_score()
  if swapped_score>=init_score:
    init_puzzle = swapped_puzzle
    init_score = swapped_score
end = time.time()

max_score = max(scores)

print(f"Maximum score reached: {max_score} in {round(end-start, 2)}s")
print(f"Maximum possible score: {max_puzzle_score}")
print(f"Therefore: {round(100*max_score/max_puzzle_score, 2)}% of pieces is correctly placed")
plt.plot(iterations, scores)
plt.title("Score as a function of the number of iterations")
plt.show()
```
